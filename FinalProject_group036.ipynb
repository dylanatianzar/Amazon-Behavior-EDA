## Data Cleaning and Wrangling

The first step in our data cleaning process was to eliminate any unnecessary columns, focusing only on those that would help us answer our research question. We carefully selected the relevant columns from the dataset to ensure our analysis was precise and targeted. Following this, we checked for any rows containing null values, as missing data could affect the integrity of our results. However, upon thorough examination, we were pleased to find that there were no null values in the columns we intended to use, allowing us to proceed with our analysis.

```python
# Clean the dataset and choose and select which features we want to use
cleaned_data = consumer_behavior[['age','Gender','Purchase_Frequency',
                                 'Personalized_Recommendation_Frequency','Browsing_Frequency',
                                 'Customer_Reviews_Importance','Review_Reliability','Review_Helpfulness',
                                 'Shopping_Satisfaction']]

# View how many null values in each of the columns
cleaned_data.isnull().sum()
```

```python
# See new shape of dataset after cleaning
data_shape = cleaned_data.shape
data_shape
```

### Encoding Columns

As part of our data wrangling process, we encoded several categorical columns to prepare the dataset for analysis. Specifically, we encoded the `Gender`, `Browsing Frequency`, and `Review Helpfulness` columns. Encoding these categorical variables into numeric values will allow us to use them effectively in our visualizations. 

For the `Gender` column, we assigned numerical values to represent different categories. 0 for Female, 1 for Male, and 2: Others. 

For the `Browsing Frequency` column, we mapped the frequency categories to a numerical scale to preserve their natural order. 0 for Rarely, 1 for Few times a week, 2 for Few times a month, and 3 for Multiple times a day.

Similarly, the `Review Helpfulness` column was converted to numerical values to facilitate quantitative analysis. 0 for No, 1 for Yes, and 2 for Sometimes.

```python
# Encode these columns: 
# Gender, Browsing Frequency and Review Helpfulness

cleaned_data.loc[:, 'Gender'] = cleaned_data['Gender'].map({'Female': 0, 'Male': 1, 'Others': 2, 'Prefer not to say': 2})
cleaned_data.loc[:, 'Browsing_Frequency'] = cleaned_data['Browsing_Frequency'].map({'Rarely': 0, 'Few times a week': 1, 'Few times a month': 2, 'Multiple times a day': 3})
cleaned_data.loc[:, 'Review_Helpfulness'] = cleaned_data['Review_Helpfulness'].map({'No': 0, 'Yes': 1, 'Sometimes': 2})
```

```python
# Remove the records with less than 15 years old
cleaned_data = cleaned_data[cleaned_data.age >= 15]

# Rename columns
cleaned_data.columns=['age','gender','purchase_frequency','recommendation_frequency','browsing_frequency','customer_reviews_importance','review_reliability','review_helpfulness','shopping_satisfaction']
cleaned_data.head()
```

Then which factors do influence shopping satisfaction? Let's examine the correlation matrix of our selected factors in relation to shopping satisfaction. The correlation matrix will reveal which variables tend to increase together or decrease as one increases.

```python
# Create copy of cleaned data
data_for_correlation = cleaned_data.copy(deep=True)

# Re-encoding any missed categorical variables
data_for_correlation['purchase_frequency'] = data_for_correlation['purchase_frequency'].map({
    'Less than once a month': 0, 'Once a month': 1, 'Few times of a month': 2, 'Once a week': 3, 'Multiple times a week': 4
})

data_for_correlation['recommendation_frequency'] = data_for_correlation['recommendation_frequency'].map({
    'No': 0, 'Yes': 1, 'Sometimes': 2
})

data_for_correlation['review_reliability'] = data_for_correlation['review_reliability'].map({
    'Never': 0, 'Rarely': 1, 'Occasionally': 2, 'Moderately': 3, 'Heavily': 4
})

# Create and plot correlation matrix
correlation_matrix = data_for_correlation.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

### Interpretation
The values in the correlation matrix correspond as such:
- **1** indicates perfect positive correlation 
- **-1** indicates perfect negative correlation
- **0** indications no correlation

Summary:
- Moderate Positive Correlations:
  - Customer reviews importance and shopping satisfaction (0.40)

- Moderate Negative Correlations:
  - Review reliability and customer reviews importance (-0.42)
  - Review reliability and shopping satisfaction (-0.26)

- Weak Correlations:
  - All other correlations indicate weak correlations between the pairs of variables


Based on the correlation matrix, we can observe that customer review importance has the strongest positive correlation with shopping satisfaction. This indicates that individuals place higher importance on customer reviews tend to be more satisfied with their shopping experience on Amazon. However, the negative correlation between review reliability and customer review importance suggests that while customers value reviews highly, they often find them unreliable on Amazon. This difference might contribute to lower shopping satisfaction scores overall. We can note that this suggests that Amazon reviews may not accurately describe the products meaning expectations are not met by how customer reviews on the products. We can infer by these findings when customers lower their expectations based on unreliable reviews, their shopping satisfaction tends to be higher. 

These results support our earlier conclusions that customer demographics, browsing frequency, and the frequency of interactions with Amazon's personalized recommendations are not correlated with shopping satisfaction on Amazon.

Let's see the distribution of the customer_reviews_importance scores.

```python
sns.countplot(data=cleaned_data,x='customer_reviews_importance')
plt.title('Customer Review Importance Distribution')
plt.xlabel('Customer Review Importance Score')
plt.ylabel('Frequency')
plt.show()
```

We can see that the distribution of the `customer_reviews_importance` scores is right-skewed. This represents that the majority of our data expresses low-moderate importance on customer reviews. 
 
We hypothesize that the `customer_reviews_importance` and `shopping_satisfaction` correlation is statistically significant. Let's check this using the Pearson's correlation coefficient. We use Pearson's correlation coefficient because we want to measure the strength and direction of the linear relationship between the two scores and provides a p-value which can be assessed for hypothesis testing. 

The p-value will be tested against the standard significance level of 0.05.

```python
from scipy.stats import pearsonr

# Calculate the Pearson correlation coefficient and p-value
corr, p_value = pearsonr(data_for_correlation['customer_reviews_importance'], data_for_correlation['shopping_satisfaction'])

print('Pearson correlation coefficient: ' + str(corr))
print('P-value: '+ str(p_value))

# Interpret the results
if p_value <= 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```
